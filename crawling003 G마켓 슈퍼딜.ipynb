```python
url = 'http://corners.gmarket.co.kr/SuperDeals'
```

```python
response = requests.get(url)
response
```

```python
dom = BeautifulSoup(response.text, 'html.parser')
dom
```

```python
elements = dom.select('#container .wrap ul li')
elements[17:]
```

```python
result = [
    el.select('li div.inner a span.title')
    for el in elements[17:]
]
result
```

```python
for el in elements[17:]:
    print(el.text.strip())
```
